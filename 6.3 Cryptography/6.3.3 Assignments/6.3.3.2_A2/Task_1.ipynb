```python
from collections import Counter
characters = [binary_cipher[i:i+8] for i in range(0, len(binary_cipher), 8)]
```

```python
N = 18
```

```python
words = []
for i in range(N):
    words.append([])

for i in range(len(characters)):
    words[i%N].append(characters[i])
```

```python
freq = []
for i in range(N):
    m = {}
    freq_ = Counter(words[i])
    
    for key in freq_.keys():
        m[key] = freq_[key] / len(words[i])
    m = dict(sorted(m.items(), key=lambda item: item[1],reverse=True))
    freq.append(m)
```

```python
the_key = ""
for i in range(N):
    space_cipher = list(freq[i].keys())[0]
    key_char = int(space_cipher,2)^32
    the_key+=chr(key_char)
    for key in freq[i].keys():
        freq[i][key] = int(key,2)^key_char
```

```python
plain_text = ""
for i in range(len(characters)):
    plain_text+=chr(freq[i%N][characters[i]])
```

```python
f = open('./plain_text.txt',"w")
f.write(plain_text)
```
