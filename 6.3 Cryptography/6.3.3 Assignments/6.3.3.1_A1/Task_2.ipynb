```python
cipher = open("./v_ciphertext_73.txt","r")
```

```python
cipher = cipher.read()
```

```python
grps = []
frequencies = []
for i in range(14):
    grps.append([])
    frequencies.append({})
```

```python
for i in range(len(cipher)):
    grps[i % 14].append(cipher[i])
```

```python
for i,grp in enumerate(grps):
    for char in grp:
        if frequencies[i].get(char,None) != None:
            frequencies[i][char]+=1
        else:
            frequencies[i][char] = 1
    for key in frequencies[i].keys():
        frequencies[i][key] = frequencies[i][key] / len(grp)
    frequencies[i] = dict(sorted(frequencies[i].items(),key=lambda item: item[1],reverse=True))
```

```python
shifts = []
for i in range(14):
    print(frequencies[i])
    h_ord = ord(list(frequencies[i].keys())[0])
    if h_ord < ord('e'):
        shifts.append(26-(ord('e')-h_ord))
    else:
        shifts.append(h_ord-ord('e'))
```

```python
decipher = ""
for i,char in enumerate(cipher):
    if ord(char) - shifts [i%7] >= ord('a'):
        decipher+=chr(ord(char)-shifts[i%7])
    else:
        decipher+=chr(ord(char)+(26-shifts[i%7]))
```

```python
from collections import Counter

def calculate_ic(text):
    """
    Calculate the Index of Coincidence (IC) for a given text.

    Args:
        text (str): The input text.

    Returns:
        float: The IC value.
    """
    freq = Counter(text)
    N = len(text)
    if N <= 1:
        return 0
    return sum(f * (f - 1) for f in freq.values()) / (N * (N - 1))

def estimate_key_length(ciphertext):
    """
    Estimate the key length by calculating IC for various key lengths.

    Args:
        ciphertext (str): The encrypted text.

    Returns:
        list: Likely key lengths with their IC values.
    """
    likely_lengths = []
    for key_len in range(1, 21):  # Test key lengths up to 20
        groups = [''.join(ciphertext[i::key_len]) for i in range(key_len)]
        avg_ic = sum(calculate_ic(group) for group in groups) / key_len
        likely_lengths.append((key_len, avg_ic))
    
    # Sort by IC value (descending) for likely key lengths
    likely_lengths.sort(key=lambda x: -x[1])
    return likely_lengths

# Example usage:  # Replace this with your ciphertext

# Estimate key length and IC values
likely_key_lengths = estimate_key_length(cipher)
print("Likely Key Lengths and IC Values:")
for length, ic in likely_key_lengths:
    print(f"Key Length: {length}, IC: {ic:.4f}")
```
